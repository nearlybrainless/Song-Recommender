```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('newcsv.csv', index_col=0)

df2 = pd.read_csv('data.csv')
```

```python
df2.head()
```

```python
df2.drop(['instrumentalness','year','explicit','mode','id','release_date','popularity','artists'],axis=1,inplace=True)
```

```python
df2.head()
```

```python
df2 =df2[['name','acousticness','danceability','duration_ms','key','liveness','loudness','speechiness','tempo','valence']]
                
```

```python
df2
```

```python
df2.set_index("name", inplace = True)
```

```python
df2.drop_duplicates(keep = False, inplace = True)
```

```python
df.drop(['energy','id','instrumentalness','mode','time_signature','type','uri'],axis=1,inplace=True)
```

```python
df.rename(columns = {'acousticenss':'acousticness'}, inplace = True)

df
```

```python
name_of_song = input("enter a song name:")

details_of_selected_song = df.loc[name_of_song]

row = df2.append(details_of_selected_song)

data_min1 = row.copy()

for column in data_min1.columns:
    data_min1[column] = (data_min1[column] - data_min1[column].min()) / (data_min1[column].max() - data_min1[column].min())
    
kmeans = KMeans(n_clusters=100)

kmeans.fit(data_min1)

data_min1['NEWCOL'] = kmeans.labels_

desired_col = data_min1.loc[name_of_song]['NEWCOL']

var1 = desired_col.drop_duplicates(keep='first')

var2 = var1.values[0]

final_data = data_min1[data_min1['NEWCOL'] == var2]

new_list = list(final_data.index.values)

print("Songs Similar to {} are".format(name_of_song))

new_list
    

```
